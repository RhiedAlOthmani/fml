Pearsonâ€™s correlation for the 2 relations

```python
from scipy.stats.stats import pearsonr #We will need the correlation function from the scientific package scipy
print(pearsonr(df_sel['average_steps'],df_sel['weight']))
print(pearsonr(df_sel['average_steps'],df_sel['height']))
```

The output gives us two numbers: r which is -0.12, and p which appears to be 0,003 but which is actually just really small. We can conclude the following:

Average steps and weight are not so strongly correlated (r = -0.12)
The correlation is highly significant (p < 0.003)

But the r of the second calculation is smaller, this means that height and average steps have a smaller colleration than with the weight
